```python
import pandas as pd


data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/mL/sun/train.csv")
data.head(10)
```

```python

lt=data['Lat'].unique()
ln=data['Lon'].unique()
print(lt)
print(ln)
copydata = data.copy()

copydata.drop(copydata.index,inplace=True)
from numpy.ma.core import count
import time
for i in lt:
  for k in ln:
    codition1=((data['Lat']==i))
    codition2=((data['Lon']==k))
    data1=data[(codition1) & (codition2)]
    n=list(data1.shape)
    print(n[0])
    print("共有",sum(data1['Temp_m'].isnull()))
    

    print(i,k)
    # print(data1.describe())
    s=data1['Temp_m'].count()
    a=data1['Temp_m'].mean()
    print(s,a)


    if(n[0]==0):
      continue
    if(s==0):
      data1['Temp_m'].fillna(value=40, inplace=True)
    else:
      data1['Temp_m'].fillna(data1['Temp_m'].mean(),inplace=True)

    copydata = pd.concat([copydata, data1], ignore_index=True)
    print("合併df來新增資料")
    print(copydata)
    # time.sleep(10)

s=copydata['Module'].unique()

print(s)

for index,n in enumerate(s):
  print(index,n)
  copydata['Module'].replace(n,index,inplace=True)

# Irradiance isnull insert maen()
copydata['Irradiance'].fillna(copydata['Irradiance'].mean(),inplace=True)
copydata['Temp'].fillna(method='ffill',inplace=True)
```

```python
copydata.isnull().sum()
```

```python
import numpy as np
import pandas as pd
## 導入視覺化套件
import matplotlib.pyplot as plt
## 導入Sklearn套件
## 導入將數據集拆成訓練集與測試集的套件
from sklearn.model_selection import train_test_split
## 導入迴歸模型套件
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np
X = pd.DataFrame(np.c_[copydata['Irradiance'], copydata['Capacity'],copydata['Irradiance_m'],copydata['Temp'],copydata['Temp_m']], columns = ['Irradiance','Capacity','Irradiance_m','Temp','Temp_m'])
y = copydata['Generation']
```

```python
## 將數據集拆成訓練集與測試集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

## 訓練與建構迴歸模型
from sklearn.linear_model import Lasso
regressor = Lasso()
regressor.fit(X_train, y_train)
```

```python
## 計算出截距值與係數值
w_0 = regressor.intercept_
w_1 = regressor.coef_

print('Interception : ', w_0)
print('Coeficient : ', w_1)

## 迴歸模型的準確度
score = regressor.score(X_test, y_test)
print('Score: ', score)
print('Accuracy: ' + str(score*100) + '%')
```

```python
## 拿訓練好的迴歸模型預測測試集資料的目標值(依變數)
y_pred = regressor.predict(X_test)

print('Predict : ', y_pred)
```
